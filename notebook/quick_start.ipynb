```python
class CustomEngine(rt.SegBase):
    @staticmethod
    def _segment(trace: str, breakpoint: int, **kwargs) -> tuple[list[tuple[int, int]], list[str]]:
        return [(0, breakpoint), (breakpoint, len(trace))], ["1", "2"]


class CustomAligner(rt.OffsetFusion):
    @staticmethod
    def fuse(segmentations: list[list[tuple[int, int]]], **kwargs) -> list[tuple[int, int]]:
        return segmentations[0]


custom_segmentor = rt.RTSeg(engines=[CustomEngine, rt.RTRuleRegex],
                            aligner=CustomAligner,
                            seg_base_unit="clause",
                            label_fusion_type="majority")
offsets, labels = custom_segmentor(rtrace, breakpoint=100)
print(offsets, labels)
```
